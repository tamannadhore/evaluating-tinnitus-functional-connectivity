```python
# libraries
import numpy as np
import nibabel as nib
import nilearn as nil
import scipy.ndimage as ndi
import matplotlib.pyplot as plt
import os
import random
```

```python
# plot 10 random averaged timeseries
os.chdir('/data/hcp_working_folder/tdhore2/subj/100307/func')

fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(30,20))
plt.subplots_adjust(hspace = 0.3)
files = ['REST1_LR', 'REST1_RL', 'REST2_LR', "REST2_RL"]

for pt in range(0,4):
    img = nib.load('rfMRI_' + files[pt] + '.nii.gz')
    # print (img.header) 
    orig_data = np.array(img.dataobj)
    mn = orig_data - orig_data.mean(axis=0)
    norm_data = mn / orig_data.std()
    plt.subplot(int('22' + str(pt+1)))
    for i in range(0,10):
        x = random.randint(0,90)
        y = random.randint(0,108)
        z = random.randint(0,90)
        data = norm_data[x, y, z, :]
        plt.xlabel ('Time')
        plt.ylabel ('Normalized Value')
        plt.plot(data.T)
    plt.title('rfMRI_' + files[pt])
plt.show()
```
