```python
# libraries
import os
import pandas as pd
import numpy as np
```

```python
# load file
os.chdir('/data/UIUC/')
data=pd.read_csv('participants.tsv',sep='\t')
with pd.option_context('display.max_rows', None,
                       'display.max_columns', None,
                       'display.precision', 3,
                       ):
    print(data)
```

```python
# sort by group
x=data.sort_values('group')
```

```python
# sort by gender within group
y=x[x['group'] == 'TNHL'] 
z=y.sort_values('sex')
with pd.option_context('display.max_rows', None,
                       'display.max_columns', None,
                       'display.precision', 3,
                       ):
    print(z)
```
